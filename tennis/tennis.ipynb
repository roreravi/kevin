2. Limpieza de Datos Básica:

Identificación y manejo de valores faltantes (NaN).
Exploración de los tipos de datos de las columnas.
Eliminación de columnas potencialmente no relevantes para un análisis inicial sencillo (por ejemplo, 'Odd_1', 'Odd_2').

```python
# Verificamos la cantidad de valores faltantes por columna
print(df.isnull().sum())

# Eliminamos las columnas 'Odd_1' y 'Odd_2' ya que no las usaremos en este ejercicio básico
df_cleaned = df.drop(columns=['Odd_1', 'Odd_2'])
print("\nDataFrame después de eliminar columnas:")
print(df_cleaned.head())

# Verificamos los tipos de datos de las columnas del DataFrame limpio
print("\nTipos de datos del DataFrame limpio:")
print(df_cleaned.dtypes)
```

sum(): Suma los valores True (que se interpretan como 1) por cada columna del DataFrame booleano resultante. Esto da como resultado una Serie que muestra la cantidad de valores faltantes en cada columna de df.

El resultado de esta operación se asigna a un nuevo DataFrame llamado df_cleaned, manteniendo el DataFrame original df sin modificaciones.

3. Consultas y Filtrado Básico:

Selección de columnas específicas (ej. 'Tournament', 'Player_1', 'Player_2', 'Winner').
Filtrado de datos basado en condiciones (ej. partidos jugados en un torneo específico, partidos ganados por un jugador en particular).


```python
# Seleccionamos algunas columnas de interés
df_subset = df_cleaned[['Tournament', 'Date', 'Player_1', 'Player_2', 'Winner']]
print("Subconjunto de columnas:")
print(df_subset.head())

# Filtramos los partidos jugados en el torneo 'Australian Open'
australian_open = df_subset[df_subset['Tournament'] == 'Australian Open']
print("\nPartidos del Australian Open:")
print(australian_open.head())

# Filtramos los partidos donde 'Player_1' fue el ganador
player1_wins = df_subset[df_subset['Winner'] == df_subset['Player_1']]
print("\nPartidos donde el Jugador 1 ganó:")
print(player1_wins.head())

# Podemos combinar condiciones (ej. partidos del Australian Open ganados por 'Roger Federer')
federer_wins_ao = australian_open[australian_open['Winner'] == 'Federer R.']
print("\nPartidos del Australian Open ganados por Roger Federer:")
print(federer_wins_ao.head())
```

4. Visualización Simple con Pandas:

Creación de gráficos básicos para entender la distribución de datos (ej. conteo de partidos por superficie).

```python
import matplotlib.pyplot as plt

# Contamos la cantidad de partidos por superficie
surface_counts = df_cleaned['Surface'].value_counts()
print("\nCantidad de partidos por superficie:")
print(surface_counts)

# Creamos un gráfico de barras
surface_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Cantidad de Partidos por Superficie')
plt.xlabel('Superficie')
plt.ylabel('Número de Partidos')
plt.xticks(rotation=45, ha='right') # Rotamos las etiquetas del eje x para mejor legibilidad
plt.tight_layout() # Ajusta los márgenes para evitar que las etiquetas se superpongan
plt.show()
# Investigar otros tipos de gráficas
```

import matplotlib.pyplot as plt: Importa la sublibrería pyplot de Matplotlib, que proporciona una interfaz para crear gráficos y visualizaciones, y la asigna el alias plt.

surface_counts = df_cleaned['Surface'].value_counts():

df_cleaned['Surface']: Selecciona la columna 'Surface' del DataFrame df_cleaned, lo que resulta en una Serie con los nombres de las superficies de cada partido.

value_counts(): Es una función de Pandas que cuenta la frecuencia de cada valor único en la Serie. Devuelve una nueva Serie donde los índices son los valores únicos de la columna 'Surface' y los valores son sus respectivas frecuencias (la cantidad de veces que aparece cada superficie).

print("\nCantidad de partidos por superficie:"): Imprime un encabezado.

print(surface_counts): Muestra la Serie surface_counts.

surface_counts.plot(kind='bar', figsize=(10, 6)):

plot(): Es una función de Pandas que se utiliza para crear gráficos a partir de Series o DataFrames. El tipo de gráfico se especifica con el argumento kind.

kind='bar': Indica que se debe crear un gráfico de barras. Los índices de la Serie surface_counts se utilizarán como las etiquetas del eje x y los valores como la altura de las barras.

figsize=(10, 6): Define el tamaño de la figura del gráfico en pulgadas (ancho=10, alto=6).

plt.title('Cantidad de Partidos por Superficie'): Establece el título del gráfico utilizando la función title() de Matplotlib.

plt.xlabel('Superficie'): Establece la etiqueta del eje x utilizando la función xlabel() de Matplotlib.

plt.ylabel('Número de Partidos'): Establece la etiqueta del eje y utilizando la función ylabel() de Matplotlib.

plt.xticks(rotation=45, ha='right'): Rota las etiquetas del eje x 45 grados hacia la derecha (ha='right') para evitar que se superpongan si son largas.

plt.tight_layout(): Ajusta automáticamente los márgenes de la subtrama para proporcionar un diseño ajustado, lo que puede ayudar a evitar que las etiquetas se recorten.

plt.show(): Muestra el gráfico creado.
